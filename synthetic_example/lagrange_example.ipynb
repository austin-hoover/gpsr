```python
import torch
from torch.autograd import grad
# define a function to optimize
def f(x):
    return x**2

x = torch.tensor(0.0, requires_grad=True)
_lambda = torch.tensor(1.0, requires_grad=True)
lr = 0.01

track = []
for i in range(100):
    x.grad = None
    _lambda.grad = None
    # get gradients
    loss = f(x) + _lambda * (x**2.0 - 1)
    z = grad(loss, (x, _lambda), create_graph=True)

    # compute the magnitude of the gradient
    h = torch.sqrt(z[0]**2 + z[1]**2)
    z2 = grad(h, (x, _lambda), create_graph=True)

    x = x - lr*z2[0]
    _lambda = _lambda - lr*z2[1]

    track.append([x, _lambda])
    print(x)


```

```python
import torch

x = torch.tensor(0, requires_grad=True, dtype=torch.float64)
y = torch.tensor(0, requires_grad=True, dtype=torch.float64)
l = torch.tensor(0, requires_grad=True, dtype=torch.float64)
lr = 0.01

track = []

def reset(tensors):
    for ele in tensors:
        ele.grad.detach_()
        ele.grad.zero_()

# min x^2+y^2 s.t x+y = 1
for i in range(1000):
    L = x**2 + y**2 + l*(1-x-y)
    L.backward()
    x.data.add_(-lr*x.grad.data)
    y.data.add_(-lr*y.grad.data)
    reset([l,x,y])
    with torch.no_grad():
        l_grad = (1-x-y)
    l.data.add_(lr*l_grad)
    reset([l,x,y])

    print(x)


print(x, y, x+y,l)
```
