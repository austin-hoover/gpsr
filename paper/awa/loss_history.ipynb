```python
import matplotlib.pyplot as plt
import torch

tkwargs = {"dtype": torch.float}
scales = [0.90, 0.95, 1.0, 1.05, 1.1]


losses = []

for scale in scales:
    tkwargs = {"dtype": torch.float}
    save_dir = f"uncertainty/ensemble_mse_scale_{scale}"
    traces = torch.load(f"uncertainty/ensemble_mse_scale_{scale}/loss_log.pt")

    losses += [torch.stack([ele[0] for ele in traces])]
losses = torch.stack(losses)
```

```python
for ele in losses:
    plt.semilogy(ele[:, 0].cpu().detach())
```
