```python
import numpy as np
import matplotlib.pyplot as plt

n_particles = [     10,     100,    1000,   10000,  100000, 1000000]

memory_peak_grad = [1.47789860e+00, 1.47761250e+00, 4.43943596e+00, 4.49498863e+01,
               2.83835270e+02, 2.90265345e+03]
memory_peak_no_grad = [np.nan,np.nan,np.nan,1.59540081, 15.59066296, 528.91836071]

fig,ax = plt.subplots(figsize=(4,3))
ax.loglog(n_particles, memory_peak_grad, ".",label="Tracking with gradients")
ax.loglog(n_particles, memory_peak_no_grad, ".",label="Tracking without gradients")
ax.legend()
ax.set_xlabel("Number of beam particles")
ax.set_ylabel("Peak memory consumption (MB)")
fig.tight_layout()
fig.savefig("memory_benchmarking.png", dpi=300)
```
