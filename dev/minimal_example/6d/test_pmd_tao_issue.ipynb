```python
from pmd_beamphysics import ParticleGroup
from pytao import Tao
from distgen import Generator
gen = Generator("test_input.yaml")
particles = gen.run()
particles.drift_to_z(z=0)
particles.write('test_beam.h5')
```

```python
pg_file = ParticleGroup('test_beam.h5')
pg_file.plot('t', 'pz')
```

```python
tao = Tao('-init test_tao.init -noplot')
tao.cmd('set global track_type = beam')
pg_tao = ParticleGroup(data=tao.bunch_data('beginning'))
pg_tao.plot('z','pz')
```

```python
tao.bunch1('end', 'z')
```
